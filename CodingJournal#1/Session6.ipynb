```python
import numpy as np

def sin(x):
    '''
    input:
        x - x value used to compute sine
    output:
        the output of sine of x.
    '''
    return np.sin(x)
```

```python
def cos(x):
    '''
    input:
        x - x value used to compute cosine
    output:
        the output of cosine of x.
    '''
    return np.cos(x)
```

```python
x_values = np.linspace(0,2*np.pi,1000)
sine_values = sin(x_values)
cosine_values = cos(x_values)
```

```python
print("x\tsin(x)\tcos(x)")
for i in range(10):
    print(f"{x_values[i]:3.3f}\t{sine_values[i]:3.3f}\t{cosine_values[i]}")
```
