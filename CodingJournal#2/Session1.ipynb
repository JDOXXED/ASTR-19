```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#exponent function
def exponent_func(a,b,x):
    return a * np.exp(-b * x) #essentially a * e^(-b * x)
```

```python
x = np.linspace(0,2*np.pi, num = 100)
y = exponent_func(5,2,x)
```

```python
plt.figure(figsize = (4,4))
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.savefig("session_1.pdf")
```
